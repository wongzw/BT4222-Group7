```python
import pandas as pd
import numpy as np
```

```python
df_tracks = pd.read_csv('tracks_features.csv')
df_tracks = df_tracks[df_tracks["year"] > 2000]
df_tracks["name"] = df_tracks["name"].apply(lambda x: x.lower())
df_tracks["album"] = df_tracks["album"].apply(lambda x: x.lower())
df_tracks["artists"] = df_tracks["artists"].apply(lambda x: x[2:-2].lower())
```
